# Cleaning World Hapiness 2015

```python
import pandas as pd
happiness2015=pd.read_csv("World_Happiness_2015.csv")
happiness2015
```

```python
#checking for the columns
happiness2015.columns
```

```python
happy=happiness2015.replace(['Economy (GDP per Capita)','Health (Life Expectancy)'],['Econ','Healthy'], inplace=True)
happy
```

```python
#Changing column names
mapping = {'Economy (GDP per Capita)': 'Economy', 'Health (Life Expectancy)': 'Health', 'Trust (Government Corruption)':'Trust','Dystopia Residual':'Dystopia'}
happiness2015.rename(columns=mapping, inplace=True) 
happiness2015.columns
```

```python
#Transforming the Economy,Family and Health Data
def label(element):
    if element > 1:
        return 'High'
    else:
        return 'Low'
factors = ['Economy', 'Family', 'Health']
factors_impact=happiness2015[factors].applymap(label)
factors_impact
```

```python
#checking for duplicates
happiness2015.duplicated()
```

```python
#checking for sum of duplicates
happiness2015.duplicated().sum()
```

```python
#checking for null values
happiness2015.isnull()
```

```python
#checking for the sum of null values
happiness2015.isnull().sum()
```

```python
#reading backto the csv file
happiness2015.to_csv('Hapiness 2015',index=False)
```

# Concatenating 2015,2016 and 2017

```python
import pandas as pd
WH_2015=pd.read_csv("World_Happiness_2015.csv")
WH_2016=pd.read_csv("Wh_2016.csv")
WH_2017=pd.read_csv("Wh_2017.csv")
```

```python
#renaming columns in World Hapiness 2016
mapping2 = {'Economy (GDP per Capita)': 'Economy', 'Health (Life Expectancy)': 'Health', 'Trust (Government Corruption)':'Trust','Dystopia Residual':'Dystopia'}
WH_2016.rename(columns=mapping2, inplace=True) 
WH_2016.columns
```

```python
WH_2017.columns
```

```python
#renaming columns in World Hapiness 2017
mapping3 = {'Economy..GDP.per.Capita.': 'Economy', 'Health..Life.Expectancy.': 'Health', 'Trust..Government.Corruption.':'Trust','Dystopia.Residual':'Dystopia','Happiness.Rank':'Happiness Rank','Happiness.Score':'Happiness Score',}
WH_2017.rename(columns=mapping3, inplace=True) 
WH_2017.columns
```

```python
WH_2015_TO_2017=pd.concat([WH_2015, WH_2016, WH_2017], ignore_index=True)
WH_2015_TO_2017
```

```python
#Checking for duplicate values
WH_2015_TO_2017.duplicated()
```

```python
#checking for null values
WH_2015_TO_2017.isnull()
```

```python
all_columns=WH_2015_TO_2017.columns
all_columns
```

```python
WH_2015_TO_2017[all_columns].fillna(value=WH_2015_TO_2017[all_columns].mean()) 
```

```python
#reading backto the csv 
WH_2015_TO_2017.to_csv('Hapiness 2015_2016_2017',index=False)
```
